```python
import jieba
import re

#jiebaè‡ªå®šä¹‰è¯åº“
# â‘ ä¸€ä¸ªä¸ªæ·»åŠ 
word_seg_list = ['é‚¹å¸‚æ˜','ä¸å†’ä¸å¤±','å¤§æ•°æ®','æœºå™¨å­¦ä¹ ','æ˜Ÿè¾°å¤§æµ·','æœ¬','åˆ äºº','å¾®ä¿¡','è¶å˜','è¹‰è·']
for i in word_seg_list:
    jieba.add_word(i) # æ·»åŠ 
#jieba.del_word("ä¸å†’") # åˆ é™¤
# ç›´æ¥åŠ è½½å­—å…¸æ–‡ä»¶
#æ ¼å¼ï¼šä¸€ä¸ªè¯å ä¸€è¡Œï¼›æ¯ä¸€è¡Œåˆ†ä¸‰éƒ¨åˆ†ï¼šè¯è¯­ã€è¯é¢‘ï¼ˆå¯çœç•¥ï¼‰ã€è¯æ€§ï¼ˆå¯çœç•¥ï¼‰ï¼Œç”¨ç©ºæ ¼éš”å¼€ï¼Œé¡ºåºä¸å¯é¢ å€’ã€‚
#file_name è‹¥ä¸ºè·¯å¾„æˆ–äºŒè¿›åˆ¶æ–¹å¼æ‰“å¼€çš„æ–‡ä»¶ï¼Œåˆ™æ–‡ä»¶å¿…é¡»ä¸º UTF-8 ç¼–ç ã€‚
#jieba.load_userdict(file_name) 
# ä¸­æ–‡åœç”¨è¯è¡¨ï¼Œä¸‹è½½åœ°å€ï¼š
# â‘ https://download.csdn.net/download/ybk233/10606306
# â‘¡1893ä¸ªï¼Œhttps://blog.csdn.net/shijiebei2009/article/details/39696571
stopword_file = 'stopword_china.txt'
stopword_list = [i.strip() for i in open(stopword_file, encoding='utf8')]
#stopword_list
```

```python

sign_list = []
cut_list = []
sign_dict = {'empty':{'ç”·':0, 'å¥³':0, 'å…¶ä»–':0},
             'not':{'ç”·':0, 'å¥³':0, 'å…¶ä»–':0},
             'all':{'ç”·':0, 'å¥³':0, 'å…¶ä»–':0}}
word_dict = {} # è¯é¢‘è®°å½•
print(sex_dict)
for i in friends:
    signature = i["Signature"]
    signature.replace(" ", "").replace("span", "").replace("class", "").replace("emoji", "")
    #rep = re.compile("1f\d.+")
    rep = re.compile("<span.*?>.*?</span>")
    signature = rep.sub("", signature.replace('\n','|'))
    # æ€§åˆ«
    cur_sex = sex_dict[str(i['Sex'])][0]
    sign_dict['all'][cur_sex] += 1
    if signature:
        sign_dict['not'][cur_sex] += 1
    else:
        sign_dict['empty'][cur_sex] += 1
        continue
    # å½“å‰ç­¾ååˆ†è¯
    sign_cut = [w for w in jieba.cut(signature)]#, cut_all=True)
    for w in sign_cut:
        w = w.strip()
        if w not in word_dict:
            word_dict[w] = {'ç”·':0,'å¥³':0,'å…¶ä»–':0, 'all':0}
        word_dict[w][cur_sex] += 1
        word_dict[w]['all'] += 1
    sign_list.append([signature, cur_sex])
    cut_list.extend(sign_cut)
    print('[%s] %s ==> %s'%(cur_sex, signature, '/'.join(sign_cut)))
#print(word_dict)
word_list = []
# å»åœç”¨è¯
for w in word_dict:
    if not w or w in stopword_list:
        #word_dict.pop(w)
        continue
    word_list.append([w, word_dict[w]])
#print(word_list)
word_list = sorted(word_list, key=lambda x:x[1]['all'], reverse=True)
print(word_list)
```
