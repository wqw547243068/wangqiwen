```python
import math
#pip install pillow
from PIL import Image

def createImg(img_dir, out_dir):
    # æ‹¼æ¥å¥½å‹å¾®ä¿¡å¤´åƒ
    x = 0
    y = 0
    imgs = os.listdir(img_dir)
    random.shuffle(imgs)
    # åˆ›å»º640*640çš„å›¾ç‰‡ç”¨äºå¡«å……å„å°å›¾ç‰‡
    newImg = Image.new('RGBA', (640, 640))
    # ä»¥640*640æ¥æ‹¼æ¥å›¾ç‰‡ï¼Œmath.sqrt()å¼€å¹³æ–¹æ ¹è®¡ç®—æ¯å¼ å°å›¾ç‰‡çš„å®½é«˜ï¼Œ
    width = int(math.sqrt(640 * 640 / len(imgs)))
    # æ¯è¡Œå›¾ç‰‡æ•°
    numLine = int(640 / width)

    for i in imgs:
        img_file = './%s/%s'%(img_dir, i)
        try:
            img = Image.open(img_file)
        except Exception:
            print('å›¾ç‰‡æ–‡ä»¶è¯»å–å¼‚å¸¸ï¼š%s'%(img_file))
            continue
        # ç¼©å°å›¾ç‰‡
        img = img.resize((width, width), Image.ANTIALIAS)
        # æ‹¼æ¥å›¾ç‰‡ï¼Œä¸€è¡Œæ’æ»¡ï¼Œæ¢è¡Œæ‹¼æ¥
        newImg.paste(img, (x * width, y * width))
        x += 1
        if x >= numLine:
            x = 0
            y += 1
    out_file = '%s/all_%s.png'%(out_dir, user)
    newImg.save(out_file)
    return out_file

# æ‹¼æ¥å¾®ä¿¡å·æœ‰å¤´åƒ
out_dir = '.'
out_file = createImg(img_dir, out_dir)
# å°†è¿™å¼ å¾—åˆ°çš„å›¾ç‰‡å‘é€åˆ°ä½ çš„å¾®ä¿¡
itchat.send_image(out_file)
# å½“ç„¶ä½ ä¹Ÿå¯ä»¥å°†è¿™å¼ å¾—åˆ°çš„å›¾ç‰‡åˆ†äº«ç»™ä½ çš„å¥½å‹
itchat.send_image(out_file, toUserName='filehelper') 
```
